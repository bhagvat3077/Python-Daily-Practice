```python
#Random no generate from 1,6 if 6 is coming then you have to sing a song
#if 1 you have to dance
import random
i=1
while True:
    num=random.randint(1,6)
    if num==1:
        print("You have to dance and this is occur after",i,"and num is:",num)
        break
    elif num==6:
        print("You have to sing a song and this is occur after",i,"and num is:",num)
        break
    else:
        print(num)
        i=i+1
```

```python
#1 to 70 : not divisible by 7
i=1
while i<=70:
    if i%7==0:
        i=i+1
        continue
    print(i,end=" ")
    i=i+1
print("***END***")    
```
