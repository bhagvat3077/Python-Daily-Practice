```python
# 9. Group words by their starting letter into a dictionary.
words = ["apple", "banana", "apricot", "blueberry", "cherry", "avocado"]

grouped = {}

for word in words:
    first_letter = word[0]  # get the first letter
    if first_letter not in grouped:
        grouped[first_letter] = []  # create a new list if key not present
    grouped[first_letter].append(word)  # add the word to the list

print(grouped)
```

```python
#11. Build a mini command-line calculator using functions and conditionals (add, subtract, multiply, divide).
def add (a,b):
    return a+b
def sub (a,b):
    return a-b
def mult (a,b):
    return a*b
def div (a,b):
    if b ==0:
        return "Cannot devided by zero"
    return a/b
def calculate():
    a = float(input("Enter a number:-"))
    op = input("Enter opeartion like +,-,*,/")
    b = float(input("Enter a number"))

    if op == "+":
        print("Addition :-",add(a,b))
    elif op =="-":
        print("Subtraction:-",sub(a,b))
    elif op == "*":
        print("Multiplication:-",mult(a,b))  
    elif op == "/":
        print("Division:-",div(a,b)) 
    else:
        print("Invalid operator")  
calculate()            
```

```python
# 12. Create a function that takes a list and returns a new list with each element doubled.
def list1(numlist):
    doublelist = []
    for i in numlist:
        doublelist.append(i*2)
    return doublelist

number = [10,20,30,40,50] 
result = list1(number)
print("After double element in doublelist:-",result)   
    
```

```python
# 14. Invert a dictionary where duplicate values are handled by storing keys in a list.
original_dict = {"a": 1, "b": 2, "c": 1, "d": 3}
inverted_dict = {}

for key, value in original_dict.items():
    if value not in inverted_dict:
        inverted_dict[value] = [key]   # create a new list for this value
    else:
        inverted_dict[value].append(key)  # append the key if value already exists

print(inverted_dict)
```

```python
# 15. Swap the first and last elements of a tuple without modifying the original tuple.
original_tuple = (10, 20, 30, 40, 50)

# Create a new tuple with first and last elements swapped
if len(original_tuple) > 1:
    new_tuple = (original_tuple[-1],) + original_tuple[1:-1] + (original_tuple[0],)
else:
    new_tuple = original_tuple  # If tuple has 0 or 1 element, no swap needed

print("Original tuple:", original_tuple)
print("New tuple:", new_tuple)
```
