```python
###String manipulation
#1 . Print each character separated by a dash -
s = "Hello"
print('-'.join(s))
```

```python
#2 Count uppercase and lowercase letters in string
s = "Hey How are you"
upper = 0
lower = 0

for ch in s:
    if ch.isupper():
        upper =upper+1
    elif ch.islower():
        lower = lower+1

print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
#3 Count number of digits and alphabet characters in string
s = "Bhagvat3077"
digits = 0
alphabets = 0

for ch in s:
    if ch.isdigit():
        digits =digits+1
    elif ch.isalpha():
        alphabets = alphabets+1

print("No of Digits:", digits)
print("No of Alphabets:", alphabets)
```

```python
# 6 Remove all spaces from a string and print new string
s = "Hello World Python"
new_string = ""

for ch in s:
    if ch != " ":       # check if character is NOT a space
        new_string += ch  # add it to the new string

print(new_string)

```

```python
# 7 relplace all spaces with "*" in string and print the result
s = "Bhagvat3077Varade"
result = ""

for ch in s:
    if ch.isdigit():
        result += "*"
    else:
        result += ch

print(result)
```

```python
# 8 print only consonant in given string
s = "Bhagvat Varade"
vowels = "aeiouAEIOU"

print("Vowels:", end=" ")
for ch in s:
    if ch in vowels:
        print(ch, end="")

print("\nConsonants:", end=" ")
for ch in s:
    if ch not in vowels and ch != " ":
        print(ch, end="")


```

```python
#Palidrom
word = "Madam"

# Convert to lowercase to ignore case
word_lower = word.lower()

# Check palindrome
if word_lower == word_lower[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```
