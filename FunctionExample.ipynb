```python
def tax(a, b=5):
    tax_amount = a * (b / 100)
    net_salary = a - tax_amount
    return net_salary


# Taking input
a = float(input("Enter your salary: "))
b = input("Enter tax percentage (press Enter for default 5%): ")

if b == "":
    result = tax(a)
else:
    result = tax(a, int(b))

print("Your Net salary is:", result)
```

```python
# calculation program
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mult(a,b):
    return a*b
def div(a,b):
    if b == 0:
        return "Cannot divided by zero"
    return a/b
def operation():
    a = float(input("Enter a number:-"))
    op = input("Enter opeartion like +,-,*,/")
    b = float(input("Enter a number"))

    if op == "+":
        print("Addition :-",add(a,b))
    elif op =="-":
        print("Subtraction:-",sub(a,b))
    elif op == "*":
        print("Multiplication:-",mult(a,b))  
    elif op == "/":
        print("Division:-",div(a,b)) 
    else:
        print("Invalid operator")    
operation()             
```
