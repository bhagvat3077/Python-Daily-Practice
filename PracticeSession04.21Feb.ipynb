```python
# 8. Convert string to dictionary of word frequency
def word_frequency(text):
    words = text.split()
    freq = {}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
result = word_frequency("apple banana apple orange banana apple")
print(result)
```

```python
# 11. Implement a function that  Compress a string (aaabbc â†’ a3b2c1)
def compress_string(s):
    result = ""
    count = 1
    
    for i in range(len(s) - 1):
        if s[i] == s[i + 1]:
            count += 1
        else:
            result += s[i] + str(count)
            count = 1
    
    result += s[-1] + str(count)
    return result
output = compress_string("aaabbc")
print(output)

```

```python
# 12 Give list of intergers , find Longest increasing consecutive elements (simple forward check logic)
numbers = [1, 2, 2, 3, 4, 1, 2, 3, 4, 5, 1]

longest = []
current = []

for i in range(len(numbers)):
    if i == 0 or numbers[i] > numbers[i-1]:
        current.append(numbers[i])
    else:
        if len(current) > len(longest):
            longest = current
        current = [numbers[i]]

# Check last sequence
if len(current) > len(longest):
    longest = current

print("Longest increasing consecutive elements:", longest)
```

```python
# 13 write a function that validate Password validation(length,uppercase,digit,special charactor)
import string

def validate_password(password):
    # Check minimum length
    if len(password) < 8:
        return False
    
    # Check for at least one uppercase letter
    has_upper = any(c.isupper() for c in password)
    
    # Check for at least one digit
    has_digit = any(c.isdigit() for c in password)
    
    # Check for at least one special character
    has_special = any(c in string.punctuation for c in password)
    
    # Return True only if all conditions are met
    return has_upper and has_digit and has_special


# Example Usage
pw = "Passw0rd!"
if validate_password(pw):
    print("Valid Password")
else:
    print("Invalid Password")
```
