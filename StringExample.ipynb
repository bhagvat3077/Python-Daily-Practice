```python
#String Method----
print("Bhagvat123".isalnum())   # True
print("Bhagvat 123".isalnum())  # False (space)
print("Bhagvat@123".isalnum())  # False (@)
print("WELCOME".isalnum())      # True
```

```python
#String Method check-isdecimal
print("12345".isdecimal())
print("123.45".isdecimal()) # false due to - dot

print("123".isdigit())    # True
print("123".isdecimal())  # True

print("²".isdigit())      # True
print("²".isdecimal())    # False isdecimal() sabse strict hai


```
