```python
# used the"count,extend,index,insert,pop,remove,reverse,sort" operation
numlist=[10,20,30,40,20,10,70,66,78,79,88,70]
print(numlist.count(20)) # count used To check frequency of an element (how often something occurs).
print(numlist.extend([777,888])) #Adds elements of another list to the end of the current list.
print(numlist)
print(numlist.index(777)) #Returns the position (index) of the first occurrence of an element
print(numlist.insert(1,100)) #Inserts an element at a specific position.
print(numlist)
print(numlist.pop(1)) #Removes and returns an element from a given index (last item by default).
print(numlist)
print(numlist.remove(777)) #Removes the first occurrence of a specified value.--POP and removed are same but differen with index and value
print(numlist)
print(numlist.reverse()) #Reverses the order of the list.
print(numlist)
print(numlist.sort()) #Sorts the list in ascending order (default)
print(numlist)
print(numlist.__reversed__)
```

```python
# 22. Merge two lists. by using for loop
list1 = [10,20,30,40,50]
list2 = [60,70,80,90,100]
list3 = []

for i in list1:
    if i not in list3:
        list3.append(i)
for i in list2:
    if i not in list3:
        list3.append(i)
print(list3)        


```
