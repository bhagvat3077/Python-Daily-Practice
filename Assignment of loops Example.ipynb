```python
#find duplicate in "aabbcdde"
#output is abd
a = "aabbbcdde"
r = ""
d = ""

for i in a:
    if i not in r:
        r = r + i
        count = 0
        for j in a:
            if i == j:
                count += 1
        if count > 1:
            d = d + i

print("duplicate characters is equal to","=", d)
```

```python
#find no of Vowels and consonent in string "Hello How are you"
text = "Hello How are you"

vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for i in text:            
    if i in vowels:
        vowel_count += 1
    elif i != " ":        # ignore spaces
        consonant_count += 1

print("Text:", text)
print("Vowels:", vowel_count)
print("Consonants:", consonant_count)

```
