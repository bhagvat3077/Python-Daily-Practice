```python
# 1. Check if a year is a leap year

year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# 2. Print numbers from 1 to N
n = int(input("Enter N: "))

for i in range(1, n + 1):
    print(i,end=" ")
```

```python
# 3. Print number pattern
rows = int(input("Enter rows: "))

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
#. 4 Take an integer from the user and reverse it using a for loop.
num = int(input("Enter number: "))
rev = 0

for _ in range(len(str(num))):
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print("Reversed number:", rev)
```

```python
# 5. Check if a Number is Palindrome
num = int(input("Enter number: "))
temp = num
rev = 0

for _ in range(len(str(num))):
    rev = rev * 10 + temp % 10
    temp //= 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# 6. Find All Factors of a Number
num = int(input("Enter number: "))

print("Factors:")
for i in range(1, num + 1):
    if num % i == 0:
        print(i,end=" ")
```

```python
# 7. Check if a Number Is a Perfect Number
#Example: 6 → 1 + 2 + 3 = 6
num = int(input("Enter number: "))
sum_div = 0

for i in range(1, num):
    if num % i == 0:
        sum_div += i

if sum_div == num:
    print("Perfect Number")
else:
    print("Not Perfect")
```

```python
# 8. Reverse a Number
num = int(input("Enter number: "))
rev = 0

for _ in range(num):
    if num == 0:
        break
    rev = rev * 10 + num % 10
    num //= 10

print("Reversed:", rev)
```

```python
# 9. Check if Number Is Duck Number
# (Number contains at least one zero, but does not start with zero)
num = int(input("Enter number: "))
temp = num
has_zero = False

for _ in range(num):
    if temp == 0:
        break
    if temp % 10 == 0:
        has_zero = True
    temp //= 10

if has_zero:
    print("Duck Number")
else:
    print("Not Duck Number")
```

```python
# 10. Check if Two Numbers Are Co-Prime
#0 and 20 are not co-prime because they share common factors greater than 1, and their GCD is 10.”
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

if gcd == 1:
    print("Co-prime")
else:
    print("Not Co-prime")
```
