```python
# 8 Count frequency of each character in a string and store it in a dictionary.
s = "banana"
freq = {}

for char in s:
    if char in freq:
        freq[char] += 1  # increment the count of this character
    else:
        freq[char] = 1   # first occurrence of this character

print(freq)
```

```python
# 9 Find the second largest number in list without sort()
largest = 0  # start with a very small number
second_largest = 0

list1 = [10,200,30,40,55,67,777,88,99,100]

for num in list1:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print("Largest:", largest)
print("Second largest:", second_largest)



```

```python
# 10 check two string are anagram
def is_anagram(str1, str2):
    # Make lowercase to ignore case
    str1 = str1.lower()
    str2 = str2.lower()
    
    # Remove spaces if needed
    str1 = str1.replace(" ", "")
    str2 = str2.replace(" ", "")
    
    # Compare sorted letters
    return sorted(str1) == sorted(str2)

# Example
s1 = "listen"
s2 = "silent"

if is_anagram(s1, s2):
    print(s1,"is Anagram ")
else:
    print("Not Anagram ")
    
```

```python
#11.2 create function that takes a list and return a dictionary with element frequency
def count_frequency(lst):
    frequency = {}
    
    for item in lst:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1
    
    return frequency


# Example usage
my_list = [1, 2, 2, 3, 1, 4, 2]
result = count_frequency(my_list)
print(result)

```

```python
#12 write a program to find all pairs in list whose sum equals to tragate number
# Example list
my_list = [2, 4, 3, 5, 7, 8, 9]
target = 7

seen = set()

print("Pairs with sum", target, ":")

for num in my_list:
    complement = target - num
    
    if complement in seen:
        print(complement, num)
    
    seen.add(num)
```

```python
#13 removed duplicate value from list while maintaining order
my_list = [1, 2, 2, 3, 4, 3, 5, 1]

seen = set()
result = []

for item in my_list:
    if item not in seen:
        result.append(item)
        seen.add(item)

print(result)
```

```python
#14 find the first non repeating char in string
string = "aabbcddeff"

frequency = {}

# Step 1: Count frequency of each character
for char in string:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Step 2: Find first character with frequency 1
for char in string:
    if frequency[char] == 1:
        print("First non-repeating character is:", char)
        break
```

```python
# 15 create a function that rotates a list by k positions
def rotate_list(lst, k):
    n = len(lst)
    k = k % n   # Handle cases where k > n
    
    return lst[-k:] + lst[:-k]


# Example
my_list = [1, 2, 3, 4, 5]
result = rotate_list(my_list, 2)

print(result)
```
