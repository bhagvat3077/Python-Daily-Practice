````python
# 7. Write a Python program to find the union of two sets. 
#    Input: `{1, 2, 3}` and `{3, 4, 5}` 
#    Output: `{1, 2, 3, 4, 5}` 

set1 = {1,2,3}
set2 = {4,5,6}
set3  = set1.union(set2)
print("The Union of set1 and set2 is:- ",set3)
````

````python
# 8. Write a Python program to find the intersection of two sets. 
#    Input: `{1, 2, 3}` and `{2, 3, 4}` 
#    Output: `{2, 3}` 
set1 = {1,2,3}
set2 = {2,3,4}
set3 = set1.intersection(set2)
print("Intersection of set1 and set2 is:- ",set3)
````

````python
# 9. Write a Python program to find the difference between two sets (elements in set A but not 
# in set B). 
#    Input: `{1, 2, 3}` - `{2, 3, 4}` 
#    Output: `{1}` 
set1 = {1,2,3}
set2 = {2,3,4}
set3 = set1.difference(set2)
print("difference between in set1 and set2 is:-",set3)
````

````python
# 10. Write a Python program to find the symmetric difference between two sets. 
#     Input: `{1, 2, 3}` and `{3, 4, 5}` 
#     Output: `{1, 2, 4, 5}` 
set1 = {1,2,3}
set2 = {3,4,5}
set3 = set1.symmetric_difference(set2)
print("Symmetric difference in set1 and set2 is:- ",set3)
````

````python
# 11. Write a Python program to check if a set is a subset of another. 
#     Input: `{1, 2}` is subset of `{1, 2, 3, 4}` → Output: `True`
set1 = {1,2}
set2 = {1,2,3,4}
set3 = set1.issubset(set2)
print("set1 is subset of set2:-",set3)
````

````python
# 12. Write a Python program to check if a set is a superset of another. 
#     Input: `{1, 2, 3}` is superset of `{2, 3}` → Output: `True` 
set1 = {1,2,3}
set2 = {2,3}
set3 = set1.issuperset(set2)
print("set1 is superset of set2:-",set3)
````

````python
# 15. Write a Python program to convert a string to a set of characters and remove all vowels. 
#     Input: `"datascience"` 
#     Output: Set of characters excluding vowels, e.g., `{'d', 't', 's', 'c', 'n'}` 
s = "datascience"
set1 = set(s)
set2 = set()
print("The string converted in SET:-",set1)
vowels = {'a','e','i','o','u','A','E','I','O','U'}
for char in set1:
    if char not in vowels:
        set2.add(char)
print("After removed a vowels in string:- ",set2)        



````
